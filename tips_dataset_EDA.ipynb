```python
# Jupyter Notebook: EDA Graphs using Seaborn

# Exploratory Data Analysis (EDA) Graphs with Seaborn
# This notebook demonstrates **univariate, bivariate, and multivariate graphs** using Seaborn on the `tips` dataset.

import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('tips')
```

```python
data.head()
```

```python
sns.histplot(data=data, x='total_bill', bins=10, kde=True, color='skyblue')
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.kdeplot(data=data, x='total_bill', shade=True, color='purple')
plt.title("Density Plot of Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("Density")
plt.show()
```

```python
sns.boxplot(data=data, y='total_bill', color='lightgreen')
plt.title("Box Plot of Total Bill")
plt.ylabel("Total Bill ($)")
plt.show()
```

```python
sns.countplot(data=data, x='day', color='orange')
plt.title("Count of Records by Day")
plt.xlabel("Day of Week")
plt.ylabel("Count")
plt.show()
```

```python
sns.scatterplot(data=data, x='total_bill', y='tip', hue='sex', style='time', size='size', palette='cool')
plt.title("Scatter Plot: Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.legend(title="Legend")
plt.show()
```

```python
sns.lineplot(data=data, x='size', y='total_bill', hue='sex', style='sex', markers=True, dashes=False)
plt.title("Line Plot: Total Bill vs Table Size")
plt.xlabel("Table Size")
plt.ylabel("Total Bill ($)")
plt.show()
```

```python
sns.boxplot(data=data, x='day', y='total_bill', hue='sex', palette='Pastel1')
plt.title("Box Plot: Total Bill by Day and Sex")
plt.xlabel("Day")
plt.ylabel("Total Bill ($)")
plt.show()
```

```python
sns.barplot(data=data, x='day', y='total_bill', hue='sex', palette='Set2', ci=None)
plt.title("Bar Plot: Average Total Bill by Day and Sex")
plt.xlabel("Day")
plt.ylabel("Average Total Bill ($)")
plt.show()
```

```python
corr = data[['total_bill','tip','size']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Heatmap of Correlation Matrix")
plt.show()
```

```python
sns.scatterplot(data=data, x='total_bill', y='tip', hue='sex', style='time', size='size', palette='cool', sizes=(50, 200))
plt.title("Multivariate Scatter Plot")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.legend(title="Legend")
plt.show()
```

```python
sns.pairplot(data=data, hue='sex', palette='Set2', diag_kind='kde')
plt.suptitle("Pair Plot of Tips Dataset", y=1.02)
plt.show()
```

```python
sns.catplot(data=data, x='day', y='total_bill', hue='sex', col='time', kind='bar', palette='Set3')
plt.suptitle("Catplot: Total Bill by Day, Sex, and Time", y=1.05)
plt.show()
```

```python
corr = data[['total_bill','tip','size']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Multivariate Heatmap: Correlation Matrix")
plt.show()
```
